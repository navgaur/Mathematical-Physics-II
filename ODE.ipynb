## **Numerical Solutions of Ordinary Differential Equations**

## Initial Value problems

Ordinary DE are the equations that have only one independent variable and can have one or more functions of that independent variable.

Example :
$$ \frac{dy}{dx} = 2 x$$
initial condition : $y(x=0) = 2$

---

##**Euler's Method**
Consider DE
$$ \frac{dy}{dx} = f(x,y)$$
initial condition $y(x_0) = y_0$.

The solution of DE is $y(x)$. Taylor expansion of $y(x)$ around the point $x=x_0$:

$$y(x) = y(x_0) + (x - x_0) \left(\frac{dy}{dx}\right)_{x=x_0}
+ \frac{(x - x_0)}{2!} \left(\frac{d^2y}{dx^2}\right)_{x=x_0} + .....$$
Considering only the first term:
$$y(x) \approx y(x_0) + (x - x_0) y'(x=x_0)$$

discreatising the system:
$$y(x_{n+1}) = y(x_n) + h y'(x_n)$$
with $h=(x_{n+1} - x_n)$

Example :
$$ \frac{dy}{dx} = 2 x$$
initial condition : $y(x=0) = 2$


```python
import numpy as np
import matplotlib.pyplot as plt

def func(x):
  return 2*x

def euler_method(ff, x0, y0 , h, n):
#  h = (xf - x0)/no_of_steps
  x,y=[],[]
  xi,yi = x0, y0
  x.append(xi)
  y.append(yi)
  for i in range(1,n+1):
    yi += h * func(xi)
    xi += h
    x.append(xi)
    y.append(yi)
  return(x,y)

h=0.05
n=50
x0=0
y0=1
xv,yv=euler_method(func, x0, y0, h, n)

print(xv[-1],yv[-1])
plt.plot(xv,yv)
plt.grid()
plt.show()
```

##**Modified Euler's Method**

$$y_{n+1} = y_n + \frac{h}{2}\left( f(x_n,y_n) + f(x_{n+1},y_{n+1}) \right) $$
where we use $y_{n+1} = y_n + f(x_n,y_n)$ to get the result:
$$y_{n+1} = y_n + \frac{h}{2}\left( f(x_n,y_n) + f(x_{n+1},y_n + h f(x_n,y_n) \right) $$

```python
import numpy as np
import matplotlib.pyplot as plt

def func(x):
  return 2*x

def mod_euler_method(ff, x0, y0 , h, n):
#  h = (xf - x0)/no_of_steps
  x,y=[],[]
  xi,yi = x0, y0
  x.append(xi)
  y.append(yi)
  for i in range(1,n+1):
    k1 = func(xi)
    k2 = func(xi+h)
    yi = yi + (h/2)*(k1+k2)
    xi = xi+ h
    x.append(xi)
    y.append(yi)
  return(x,y)

h=0.05
n=50
x0=0
y0=1
xv,yv=mod_euler_method(func, x0, y0, h, n)

print(xv[-1],yv[-1])
plt.plot(xv,yv)
plt.grid()
plt.show()
```
